```python
DATES = open("/content/gdrive/MyDrive/Colab/COMS6998EMODS/dates.txt").readlines()
DATES = [date.split("\n")[0] for date in DATES]
DATES
```

```python
import pandas as pd
import json
import glob
import ast
from tqdm.auto import tqdm

state_code = pd.read_csv('/content/gdrive/MyDrive/Colab/COMS6998EMODS/state_code.csv')
states = state_code['code'].tolist()

MONTH = {'08':'Aug_2020', '09':'Sep_2020', '10':"Oct_2020", '11':'Nov_2020', '12':'Dec_2020'}

HOME_DIR = '/content/gdrive/MyDrive/Colab/COMS6998EMODS/Hyderated_Tweets_Hate_Indicator'
DEST_DIR = '/content/gdrive/MyDrive/Colab/COMS6998EMODS/Tweets_States_hate_score/'


cnt = 0
for state in tqdm(states):
  df_list = []
  dates = []

  daily_list = []
  for date in DATES:
    month = MONTH[date.split("-")[1]]
    csv_path = os.path.join(HOME_DIR, month, date+'.csv')
    daily = pd.read_csv(csv_path, engine='python')

    daily['state'] = daily['state'].str.replace(" ", "")
    daily = daily[daily['state'] == state]
    
    row, col = daily.shape
    cnt += row
    for _ in range(row):
      dates.append(date)
      
    df_list.append(daily)

  result = pd.concat(df_list)
  result['date'] = dates
  result = result.drop(columns=['withheld_in_countries', 'withheld_scope', 'withheld_copyright'])
  result.to_csv(DEST_DIR + state + ".csv", index=False)

print(cnt)
```
