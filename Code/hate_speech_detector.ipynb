```python
import pandas as pd
import os
import glob
# from pysentimiento.sentiment import load_tass_datasets, id2labeltass
from pysentimiento import create_analyzer
from pysentimiento.preprocessing import preprocess_tweet
from tqdm.auto import tqdm

HOME_DIR = '/content/gdrive/MyDrive/Colab/COMS6998EMODS/Aug_hydrate/CSV/*'
DEST_DIR = '/content/gdrive/MyDrive/Colab/COMS6998EMODS/Aug_hydrate/Hate/'   # create folder named Hate 
files = glob.glob(HOME_DIR)
# files = ['/content/gdrive/MyDrive/Colab/COMS6998EMODS/hydrate_tweet_csv/2020-07-31.csv']

hate_speech_analyzer = create_analyzer(task="hate_speech", lang="en")

dir_list = os.listdir(DEST_DIR)
print(dir_list)

for fpath in tqdm(files):

  name = fpath.split("/")[-1]
  if name in dir_list:
    continue
  df = pd.read_csv(fpath)

  # drop row whose row['text'] is empty
  df = df[df['full_text'].notna()]
  df = df[df['state'].notna()]

  texts, labels, scores = [], [], []
  for index, row in df.iterrows():
    texts.append(preprocess_tweet(row['full_text']))

  print(fpath.split("/")[-1])

  for sent in tqdm(texts):
    
    t = hate_speech_analyzer.predict(sent)
    # max_key = max(t.probas, key=t.probas.get)
    max_value = max(t.probas.values())
    scores.append(max_value)
    # if max_value > 0.03:
    #   labels.append(1)
    # else:
    #   labels.append(0)

  # df['label'] = labels
  df['score'] = scores
  df.to_csv(DEST_DIR + fpath.split("/")[-1], index=False)
  # hate = df[df['label'] == 1]
  # hate.to_csv('/content/gdrive/MyDrive/Colab/COMS6998EMODS/selected_hate_speech/'+ fpath.split("/")[-1], index=False)

```
